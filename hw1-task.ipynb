```python
# (3) посчитать число женщин с университетским образованием
# SEX (1 = male; 2 = female). 
# EDUCATION (1 = graduate school; 2 = university; 3 = high school; 4 = others). 

result = df[ (df['SEX'] == 2) & (df['EDUCATION'] == 2) ]
result_count = len(result.index)
print('Число женщин с университетским образованием: {:d}'.format(result_count) ) 
```

```python
# (4) Сгрупировать по "default payment next month" и посчитать медиану для всех показателей начинающихся на BILL_ и PAY_

col_for_group = 'default.payment.next.month'
cols_list = list()
for cl in df.columns:
    if cl.startswith("BILL_") | cl.startswith("PAY_"):
        cols_list.append(cl)

cols_list.append(col_for_group)

#print(result_list)

print('Сгрупировать по <default.payment.next.month> и посчитать медиану для всех показателей начинающихся на BILL_ и PAY_')
result = df[cols_list].groupby(col_for_group).median()
result.head()
```

```python
# (5) постройте сводную таблицу (pivot table) по SEX, EDUCATION, MARRIAGE

# SEX (1 = male; 2 = female). 
# EDUCATION (1 = graduate school; 2 = university; 3 = high school; 4 = others).

df.pivot_table(values='MARRIAGE', index=['SEX','EDUCATION'], aggfunc='count')
```

```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

def class_limit(limit):
    cl=''
    if limit<=10000:
        cl='A'
    elif limit<=100000:
        cl='B'
    elif limit<=200000:
        cl='C'
    elif limit<=300000:
        cl='D'
    elif limit<=700000:
        cl='E'
    else:
        cl='F'
    return cl

print('Классификация кредитного лимита (A..F):')
df[['lb_class','LIMIT_BAL']].head(10)
```

```python
# (7) построить распределение LIMIT_BAL (гистрограмму)
import matplotlib.pyplot as plt
```

```python
df['LIMIT_BAL'].hist(bins=30)
plt.title("Распределение кредитного лимита");
```

```python
# (8) построить среднее значение кредитного лимита для каждого вида образования 
# и для каждого пола
# график необходимо сделать очень широким (на весь экран)

_, ax_plots = plt.subplots(figsize=(20, 6))

df.pivot_table(values='LIMIT_BAL', index=['SEX', 'EDUCATION'], aggfunc='mean').plot(
    kind='bar', ax=ax_plots
)
plt.title("Среднее значение кредитного лимита для сочетаний пола и образования")
plt.show()
```

```python
# (9) построить зависимость кредитного лимита и образования только для одного из полов

# SEX (1 = male; 2 = female). 

_, ax_plots = plt.subplots(figsize=(20, 6))

result = df[ (df['SEX'] == 1)]           
result.pivot_table(values='LIMIT_BAL', index=['EDUCATION'], aggfunc='mean').plot(
    kind='bar', ax=ax_plots
)
plt.title("Среднее значение кредитного лимита по образованию (мужчины)")
plt.show()
```
